```python
with open("input_3.txt", "r") as r:
    tx = r.readlines()
    r.close()

tx = [x.strip('\n') for x in tx]
```

```python
#part 1
def calc_joltage(x):
    high = max(x)
    loc = x.index(high)
    if loc == 0:
        after = x[loc+1:]
        joltage = high+max(after)
        return joltage
    if loc < 99 and loc > 0:
        before = x[:loc]
        after = x[loc+1:]
        test1 = high+max(after)
        test2 = max(before)+high
        joltage = (max([test1,test2]))
        return joltage
    elif loc == 99:
        x = x.replace(high,'0')
        next_high = max(x)
        joltage = next_high+high
        return joltage
```

```python
jolts = []
for bank in tx:
    jolts.append(calc_joltage(bank))

sum([int(x) for x in jolts])
```

```python
# part 2
def max_string(x_string, start_ind, last_ind):
    if last_ind !=0:
        new_ind = x_string.index(max(x_string[start_ind:last_ind]))
        high = x_string[start_ind:last_ind][new_ind]
    else:
        new_ind = x_string.index(max(x_string[start_ind:]))
        high = x_string[start_ind:][new_ind]
    x_string = x_string[new_ind+1:]
    return high, x_string
```

```python
total=[]
for x_string in tx:
    nums = []
    max_length = 12
    start_ind = 0
    last_ind = -11
    for i in range(12):
        high, x_string = max_string(x_string, start_ind, last_ind)
        last_ind = last_ind+1
        #print(last_ind)
        nums.append(high)
    
    #print(''.join(nums))
    total.append(int(''.join(nums)))
```
