```python
with open("input_3.txt", "r") as r:
    tx = r.readlines()
    r.close()

tx = [x.strip('\n') for x in tx]
```

```python
def calc_joltage(x):
    high = max(x)
    loc = x.index(high)
    if loc == 0:
        after = x[loc+1:]
        joltage = high+max(after)
        return joltage
    if loc < 99 and loc > 0:
        before = x[:loc]
        after = x[loc+1:]
        test1 = high+max(after)
        test2 = max(before)+high
        joltage = (max([test1,test2]))
        return joltage
    elif loc == 99:
        x = x.replace(high,'0')
        next_high = max(x)
        joltage = next_high+high
        return joltage
```

```python
jolts = []
for bank in tx:
    jolts.append(calc_joltage(bank))

sum([int(x) for x in jolts])
```
