```python
with open("input_5.txt", "r") as r:
    tx = r.readlines()
    r.close()
```

```python
ranges_list = tx[:182]
ingredients = tx[183:]
```

```python
range_dict = {}
for i in ranges_list:
    low,high = i.split('-')
    low = int(low)
    high = int(high)
    range_dict[i] = range(low,high)
```

```python
fresh = []
for ranges in range_dict.values():
    for ingredient_id in ingredients:
        if int(ingredient_id) in ranges:
            fresh.append(ingredient_id)
```

```python
print(len(set(fresh)))
```

```python
#part 2
range_list = []
for i in ranges_list:
    low,high = i.split('-')
    low = int(low)
    high = int(high)
    range_list.append((low,high))
```

```python
range_list.sort(reverse=False)
```

```python
current_start,current_end = range_list[0]
unique_ids = 0
for rang in range_list[1:]:
    new_start = rang[0]
    new_end = rang[1]
    if new_start<=current_end:
        current_end = max(current_end, new_end)
    else:
        unique_ids+=(current_end - current_start)

        current_start=new_start
        current_end=new_end

unique_ids += (current_end - current_start)
```
