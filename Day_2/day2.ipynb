```python
with open("input_2.txt", "r") as r:
    tx = r.readlines()
    r.close()

tx = tx[0].split(',')
```

```python
#part 1
def pattern_find(x):
    first_id, last_id = x.split('-')
    errors = []
    for i in range(int(first_id), int(last_id)+1):
        i=str(i)
        length = len(i)
        if len(i)==2:
            if i.count(i[0])==len(i):
                errors.append(i)
        elif len(i)>2 and len(i)%2==0:
            if i[:length//2]==i[length//2:]:
                errors.append(i)
    return errors
```

```python
total_errors = []
for pattern in tx:
    errors = pattern_find(pattern)
    [total_errors.append(x) for x in errors]

sum([int(x) for x in total_errors])
```

```python
def pattern_find_b(x):
    first_id, last_id = x.split('-')
    errors = []
    for i in range(int(first_id), int(last_id)+1):
        i=str(i)
        length = len(i)
        if len(i)>=2:
            if i.count(i[0])==len(i):
                errors.append(i)
        elif len(i)==4:
            if i[:2]==i[3:]:
                errors.append(i)
        elif len(i)==6:
            if i[:2]==i[2:4]==i[4:]:
                errors.append(i)
        elif len(i)==8:
            if i[:2]==i[2:4]==i[4:6]==i[6:]:
                errors.append(i)
        elif len(i)==10:
            if i[:2]==i[2:4]==i[4:6]==i[6:8]==i[8:]:
                errors.append(i)
        elif len(i)==9:
            if i[:3]==i[3:6]==i[6:]:
                errors.append(i)
        elif len(i)>2 and len(i)%2==0:
            if i[:length//2]==i[length//2:]:
                errors.append(i)
    return errors
```

```python
total_errors = []
for pattern in tx:
    errors = pattern_find_b(pattern)
    [total_errors.append(x) for x in errors]

sum([int(x) for x in total_errors])
```
