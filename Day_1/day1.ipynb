```python
with open("input_1.txt") as r:
    tx = r.readlines()
```

```python
tx = [x.strip("\n") for x in tx]
```

```python
# part 1

def calc(x, value):
    if x.startswith('L'):
        value = value - int(x.lstrip('L'))
    elif x.startswith('R'):
        value = value + int(x.lstrip('R'))
    if value<0:
        value = 100-abs(value)%100
    return value%100

value = 50
count = 0
for turn in tx:
    value = calc(turn,value)
    if value == 0:
        count+=1
print(count)   
```

```python
# part 2
def bit_calc(x, value):
    turn_count=0
    if x.startswith('L'):
        for i in range(int(x.lstrip('L'))):
            value = value-1
            if value == 0:
                turn_count+=1
            value = value%100
    elif x.startswith('R'):
        for i in range(int(x.lstrip('R'))):
            value = value+1
            if value == 100:
                turn_count+=1
            value = value%100

    return value, turn_count
```

```python
value = 50
count = 0
for turn in tx:
    value, turn_count = bit_calc(turn, value)
    count+=turn_count

print(count)
```
